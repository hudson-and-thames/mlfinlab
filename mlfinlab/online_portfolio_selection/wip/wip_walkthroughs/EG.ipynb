Exponential Gradient

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlfinlab.online_portfolio_selection import *

```

```python
stock_prices = pd.read_csv('../../tests/test_data/stock_prices.csv', parse_dates=True, index_col='Date')
stock_prices = stock_prices.dropna(axis=1)
```

```python
multiplicative_update = EG(eta=0.05, update_rule='MU')
gradient_projection = EG(eta=0.05, update_rule='GP')
expectation_maximization = EG(eta=0.05, update_rule='EM')
multiplicative_update.allocate(stock_prices, resample_by='M')
gradient_projection.allocate(stock_prices, resample_by='M')
expectation_maximization.allocate(stock_prices, resample_by='M')
```

```python
multiplicative_update.all_weights
```

```python
plt.figure(figsize=(20,10))
plt.title('EG: Multiplicative Update Rule', fontsize=40)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Portfolio Returns', fontsize=20)
plt.plot(multiplicative_update.portfolio_return)
plt.savefig('EG_MU.png')
```

```python
plt.figure(figsize=(20,10))
plt.title('EG: Gradient Projection Update Rule', fontsize=40)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Portfolio Returns', fontsize=20)
plt.plot(gradient_projection.portfolio_return)
plt.savefig('EG_GP.png')
```
