Universal Portfolio

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlfinlab.online_portfolio_selection import *

```

```python
stock_prices = pd.read_csv('../../../tests/test_data/stock_prices.csv', parse_dates=True, index_col='Date')
stock_prices = stock_prices.dropna(axis=1)
```

```python
up10 = UP(number_of_experts=10)
up10.allocate(stock_prices)
up1000 = UP(number_of_experts=1000)
up1000.allocate(stock_prices)
```

```python
up10.all_weights
up1000.all_weights
```

```python
plt.figure(figsize=(20,10))
plt.title('UP with 10 and 1000 Experts', fontsize=40)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Portfolio Returns', fontsize=20)
plt.plot(up10.portfolio_return)
plt.plot(up1000.portfolio_return)
plt.savefig('images/up10and1000.png')

```
